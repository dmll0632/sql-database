```python
import pandas as pd
```

```python
path1 = "Data Frames/ai_data.csv"
df_ai_merg = pd.read_csv(path1)
pd.set_option('display.max_columns', None)
df_ai_merg
```

```python
path2 = "Data Frames/ai_invest.csv"
df_ai_invest = pd.read_csv(path2)
pd.set_option('display.max_columns', None)
df_ai_invest
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Set the positions and width for the bars
bar_width = 0.35
positions = np.arange(len(df_ai_invest))

# Plot for GDP per capita
fig, ax1 = plt.subplots(figsize=(12, 8))

ax1.bar(positions, df_ai_invest['gdp_per_capta'], width=bar_width, label='GDP per Capita')

ax1.set_xlabel('Country')
ax1.set_ylabel('GDP per Capita')
ax1.set_title('GDP per Capita by Country')
ax1.set_xticks(positions)
ax1.set_xticklabels(df_ai_invest['country'], rotation=90)
ax1.legend()
```

```python
# Plot for Government Strategy and Commercial
fig, ax2 = plt.subplots(figsize=(12, 8))

bar_width = 0.35
ax2.bar(positions - bar_width/2, df_ai_invest['government_strategy'], width=bar_width, label='Government Strategy')
ax2.bar(positions + bar_width/2, df_ai_invest['commercial'], width=bar_width, label='Commercial')

ax2.set_xlabel('Country')
ax2.set_ylabel('Values')
ax2.set_title('Government Strategy and Commercial by Country')
ax2.set_xticks(positions)
ax2.set_xticklabels(df_ai_invest['country'], rotation=90)
ax2.legend()

# Show the plots
plt.show()
```
